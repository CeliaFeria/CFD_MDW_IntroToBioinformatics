```r
#glue two dataframe

two_together <- cbind(HDI_orde, GDP_orde)

two_together
```

```r
#eliminate column
two_together <- cbind(HDI_orde, GDP_orde[,2:3])
head(two_together)
tail(two_together)
```

```r
#scatterplot
library(ggplot2)
#ggplot(two_together)
ggplot(two_together, aes(x = GDP.per.capita, y = HDI)) + geom_point(size = 4, aes(color = CCAA))
```

```r
options(repr.plot.width = 15, repr.plot.height = 10, repr.plot.res = 100)
theme_set(theme_bw())
ggplot(two_together, aes(x = GDP.per.capita, y = HDI)) + geom_text(aes(label=Province, color=CCAA), check_overlap=T)
```

```r
HDI_mean <- aggregate(two_together$HDI, list(two_together$CCAA), FUN = mean)
GDP_mean <- aggregate(two_together$GDP.per.capita, list(two_together$CCAA), FUN = mean)
HDI_std <- aggregate(two_together$HDI, list(two_together$CCAA), FUN = sd)
GDP_std <- aggregate(two_together$GDP.per.capita, list(two_together$CCAA), FUN = sd)


HDI_std
```

```r
mean_std <- data.frame(CCAA = HDI_mean[,1], HDI_mean= HDI_mean[,2], GDP_mean = GDP_mean[,2], HDI_std= HDI_std[,2], GDP_std=GDP_std[,2])
mean_std
```

```r
#replace NAs with zeros
mean_std[4:5][is.na(mean_std[4:5])] <- 0
mean_std
```

```r
ggplot(two_together, aes(x = GDP.per.capita, y = HDI)) + geom_point(size = 4, aes(color = CCAA)) + geom_point(data = mean_std, aes(x = mean_std[,3], y = mean_std[,2])) 
```

```r
p <- (ggplot(two_together, aes(x = GDP.per.capita, y = HDI)) + 
geom_point(size = 4, aes(color = CCAA)) + 
geom_point(data = mean_std, aes(x = mean_std[,3], y = mean_std[,2])) +
geom_errorbar(data= mean_std, mapping=aes(x = mean_std[,3], y = mean_std[,2], ymin= mean_std[,2] - mean_std[,4], ymax= mean_std[,2] + mean_std[,4])))
p
```
